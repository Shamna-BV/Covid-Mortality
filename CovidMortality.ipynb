```python
# Define a mapping function
def convert_age(value):
    if value.startswith('>'):  # Example: '>60' -> 60
        return int(value[1:])
    elif '-' in value:  # Example: '50-60' -> (50+60)/2
        low, high = map(int, value.split('-'))
        return (low + high) / 2
    else:  # Numeric strings like '1'
        return int(value)

# Apply the function to the Age column
covid['Age'] = covid['Age'].apply(convert_age)

covid
```

```python
# Load sample data
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("penguins").dropna()

# Compute correlation
corr_matrix = df.select_dtypes(include=['number']).corr()

# Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()

# High correlation pairs
high_corr = corr_matrix.unstack().sort_values(ascending=False)
print(high_corr[(high_corr > 0.8) & (high_corr < 1.0)])
```

```python
x= covid.iloc[:, :-1].values
y= covid.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier= DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(x_train, y_train)
```

```python
y_pred= classifier.predict(x_test)
from sklearn.metrics import accuracy_score
 # evaluate predictions
accuracy = accuracy_score(y_test, y_pred)
print('Decision Tree(entropy) Accuracy: %.2f' % (accuracy*100))
```
